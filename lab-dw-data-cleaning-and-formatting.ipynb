## Exercise 2: Cleaning invalid Values

The dataset contains columns with inconsistent and incorrect values that could affect the accuracy of our analysis. Therefore, we need to clean these columns to ensure that they only contain valid data.

Note that this exercise will focus only on cleaning inconsistent values and will not involve handling null values (NaN or None).

*Hint*:
- *Gender column contains various inconsistent values such as "F", "M", "Femal", "Male", "female", which need to be standardized, for example, to "M" and "F".*
- *State abbreviations be can replaced with its full name, for example "AZ": "Arizona", "Cali": "California", "WA": "Washington"*
- *In education, "Bachelors" could be replaced by "Bachelor"*
- *In Customer Lifetime Value, delete the `%` character*
- *In vehicle class, "Sports Car", "Luxury SUV" and "Luxury Car" could be replaced by "Luxury"*

```python
df_car.columns
```

```python
#check which unique entries there are
df_car.gender.unique()

#make a dictionary for a map
gender_mapping = {"female": "female", "F":"female", "Femal":"female", "f": "female", "M" : "male", "Male" : "male"}

#use map
df_car["gender"] = df_car["gender"].replace(gender_mapping)

display(df_car["gender"].unique)
```

```python
df_car.state.unique()

#make a dictionary for a map
state_mapping = {"AZ": "Arizona", "WA":"Washington", "Cali":"California"}
df_car["state"] = df_car["state"].replace(state_mapping)
display(df_car["state"].unique())
```

```python
df_car.education.unique()
#make a dictionary for a map
education_mapping = {"Bachelors": "Bachelor"}
df_car["education"] = df_car["education"].replace(education_mapping)
display(df_car["education"].unique())
```

```python
df_car.customer_lifetime_value.unique()
df_car.customer_lifetime_value = df_car.customer_lifetime_value.str.strip('%')
display(df_car.customer_lifetime_value.unique())
```

```python
df_car.vehicle_class.unique()
class_mapping = {"Luxury SUV": "Luxury", "Luxury Car":"Luxury", "Sports Car":"Luxury"}
df_car["vehicle_class"] = df_car["vehicle_class"].replace(class_mapping)
display(df_car["vehicle_class"].unique())
```

```python

# Extract the middle value using split and cast to integer.
def extract_middle_value(value):
    if isinstance(value, str):
        parts = value.split('/')  # Split the string by '/'
        return int(parts[1]) if len(parts) > 1 else 0  # Return the middle value if it exists, otherwise return 0.
    else:
        return 0
# Apply the function to convert the entire column.
df_car['number_open_complaints'] = df_car['number_open_complaints'].apply(extract_middle_value)

#test
display(df_car['number_open_complaints'].head())
```
